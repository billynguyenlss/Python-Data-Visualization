```python
import matplotlib.animation as animation
import seaborn as sns

# step
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)
X = [x1,x2,x3,x4]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharey = True)
sns.set_style(style='whitegrid')
ax = [ax1, ax2, ax3, ax4]

bins1 = np.arange(-7.5, 2.5, 0.2)
bins2 = np.arange(0, 10, 0.2)
bins3 = np.arange(7, 17, 0.2)
bins4 = np.arange(12, 22, 0.2)
bins = [bins1, bins2, bins3, bins4]

axis1 = [-7.5, 2.5, 0, 0.6]
axis2 = [0, 10, 0, 0.6]
axis3 = [7, 17, 0, 0.6]
axis4 = [12, 22, 0, 0.6]
axis = [axis1, axis2, axis3, axis4]

titles = ['x1 norm', 'x2 gamma', 'x3 exponential', 'x4 uniform']
colors = ['blue','orange','green','red']

# define update function
def update_plot(curr):
    if curr == 100:
        a.event_source.stop()
    # update subplot
    for i in range(len(ax)):
        ax[i].cla()
        ax[i].hist(X[i][:curr*100], normed=True,bins = bins[i], color = colors[i])
        ax[i].axis(axis[i])
        ax[i].set_title(titles[i])
        ax[i].set_xlabel('Value')
        ax[i].set_ylabel('Frequency')
        ax[i].annotate('n = {}'.format(curr), [3,27])
    plt.tight_layout()

a = animation.FuncAnimation(fig, update_plot, interval=100, repeat=True)
```
